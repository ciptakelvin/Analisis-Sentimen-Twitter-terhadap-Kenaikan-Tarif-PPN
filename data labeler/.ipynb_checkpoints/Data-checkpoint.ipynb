```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sb

#Data Cleaner
data=pd.read_csv('2023-02-28 19.36.57.730834 PPN Naik.csv')
data=data.drop_duplicates(subset=["Tweet"])

data=data.astype({
    'Id':'int64',
    'Datetime':'datetime64',
    'View Count':'int32',
    'Reply Count':'int32',
    'Retweet Count':'int32',
    'Like Count':'int32',
    'Quote Count':'int32',
    'Mentioned Users Count':'int32',
    'User Verified':'bool',
    'User Followers Count':'int32',
    'User Statuses Count':'int32',
    'User Verified':'bool'
})

data["Year"]=data["Datetime"].dt.year
data["Month"]=data["Datetime"].dt.month
data["Day"]=data["Datetime"].dt.day

data["Kota"]=[str(i).split(",")[0] for i in data["Place"]]
data.fillna(0)

data=data[(data["Datetime"]<datetime(2022,12,31)) & (data["Datetime"]>datetime(2020,5,1))]
data.to_csv('data_clean.csv')
data
```

```python
import random
from IPython.display import clear_output

#Labeler
labeled_data=data.copy()

while True:
    location=random.randint(0,len(data))
    print(labeled_data.iloc[location,:]["Tweet"])
    if str(labeled_data.iloc[location,5])=="nan":
        print("Input Sentiment 1:negative 2:neutral 3:positive 0:exit")
        new_sentiment=input()
        if new_sentiment=="0":
            break;
        labeled_data.iloc[location,5]=new_sentiment
        clear_output()
    else:
        continue

labeled_data[labeled_data["Sentiment Label"].notnull()].to_csv('labeled.csv')
labeled_data[labeled_data["Sentiment Label"].notnull()]
```
